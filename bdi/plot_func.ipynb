```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
ls = [i/10 for i in range(-100, 101)]
```

```python
def sigmoid(x):
    return [1 / (1 + math.exp(-i)) for i in x]
fig = plt.figure()
plt.axhline(y=0, xmin=-10, xmax=10, c='gray')
plt.axvline(x=0, ymin=10, ymax=-10, c='gray')
plt.plot(ls, sigmoid(ls))
plt.axis([-10, 10, -0.1, 1.1]) 
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.grid()
```

```python
def tanh(x):
    return [((math.exp(2 * i) - 1) / (math.exp(2 * i) + 1)) for i in x]


plt.axhline(y=0, xmin=-10, xmax=10, c='gray')
plt.axvline(x=0, ymin=10, ymax=-10, c='gray')
plt.plot(ls, tanh(ls))
plt.axis([-10, 10, -1.1, 1.1]) 
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.grid()
```

```python
def leakyrelu(x):
    out = []
    for i in x:
        if i >= 0:
            out.append(i)
        else:
            out.append(i * 0.2)
    return out


plt.axhline(y=0, xmin=-10, xmax=10, c='gray')
plt.axvline(x=0, ymin=10, ymax=-10, c='gray')
plt.plot(ls, leakyrelu(ls))
plt.axis([-10, 10, -2, 10]) 
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.text(-9, 1, '$\mathtt{\lambda = 0.2}$', fontsize=23)
plt.grid()
```

```python
def mse(x):
    return [i ** 2 for i in x]

plt.axhline(y=0, xmin=-10, xmax=10, c='gray')
plt.axvline(x=0, ymin=10, ymax=-10, c='gray')
plt.plot(ls, mse(ls))
plt.axis([-10, 10, 0, 100]) 
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.grid()
```
